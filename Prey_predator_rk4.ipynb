```python
import matplotlib.pyplot as plt
import math
%matplotlib notebook
#function
dxdt=lambda t,x,y:x*(0.1-0.0014*y)
dydt=lambda t,x,y:y*(0.08*x-0.0012)
#initialization
ti=0;tf=15;h=1
n =int((tf-ti)/h)
t=0;x=4;y=9
#Visualization of code initialization
print('t  \t\t\t x \t\t y ')
print('%f \t %f \t %f '%(t,x,y))
t_plot=[];x_rk4=[];y_rk4=[]
#RK4 Method
for i in range(1,n+1):
    t_plot.append(t);x_rk4.append(x);y_rk4.append(y)
    #rk4 method
    k1=h*dxdt(t,x,y)
    l1=h*dydt(t,x,y)
    k2=h*dxdt(t+0.5*h,x+k1*0.5,y+l1*0.5)
    l2=h*dydt(t+0.5*h,x+k1*0.5,y+l1*0.5)
    k3=h*dxdt(t+0.5*h,x+k2*0.5,y+l2*0.5)
    l3=h*dydt(t+0.5*h,x+k2*0.5,y+l2*0.5)
    k4=h*dxdt(t+h,x+k3,y+l3)
    l4=h*dydt(t+h,x+k3,y+l3)
    
    #Calculate new x and y estimation
    x=x+(1/6)*(k1+2*k2+2*k3+k4)
    y=y+(1/6)*(l1+2*l2+2*l3+l4)
    #Increase x by step size to calculate next y extimate
    t=t+h
    print('%f \t %f \t %f'%(t,x,y))
    
#visulation code
t_plot.append(t);x_rk4.append(x);y_rk4.append(y)
plt.plot(t_plot,x_rk4,'bo-',t_plot,y_rk4,'yo-')
plt.xlabel('time (Day)')
plt.ylabel('Prey and predator population')
plt.legend(['Prey population','predator populaton'])
plt.grid()
plt.show()
plt.savefig('Pry_predator_rk.png')
t=t_plot
pre=x_rk4
predator=y_rk4
```
