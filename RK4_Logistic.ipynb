```python
import matplotlib.pyplot as plt
import math
%matplotlib notebook
#function
f=lambda x:(10**6)/(1+9*math.exp(19.8-x/100))
dy=lambda x,y: (10**-2*y)-(10**-8*y**2)
#initialization
xi=1980;xf=2020;h=5
n =int((xf-xi)/h)
x=1980;y=10**5
#Visualization of code initialization
print('x   \t\t\t y \t f(x) \t\t |f(x)-y|')
print('%f \t %f \t %f \t %f'%(x,y,f(x),abs(f(x)-y)))
x_plot=[];y_rk4=[];y_exact=[]
#RK4 Method
for i in range(1,n+1):
    x_plot.append(x);y_rk4.append(y);y_exact.append(f(x))
    #rk4 method
    k1=h*dy(x,y)
    k2=h*dy(x+h*0.5,y+k1*0.5)
    k3=h*dy(x+h*0.5,y+k2*0.5)
    k4=h*dy(x+h,y+k3)
    #Calculate new y estimation
    y=y + 1 / 6 * (k1+2*k2+2*k3+k4)
    #Increase x by step size to calculate next y extimate
    x=x+h
    print('%f \t %f \t %f \t %f'%(x,y,f(x),abs(f(x)-y)))
    
#visulation code
x_plot.append(x);y_rk4.append(y);y_exact.append(f(x))
plt.plot(x_plot,y_rk4,'go-',x_plot,y_exact,'r--')
plt.xlabel('Time (Year)');plt.ylabel('Population')
plt.legend(["Numerical solution(Rk4 Method)","Exact Solution"])
plt.grid()
plt.savefig('Rk_logistic.png')
safa=x_plot
safa_exact=y_exact
safa_rk=y_rk4
```
