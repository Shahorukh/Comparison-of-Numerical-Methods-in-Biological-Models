```python
#function
%matplotlib notebook
import math
import matplotlib.pyplot as plt
f=lambda x:(10**6)/(1+9*math.exp(19.8-x/100))
dy=lambda x,y: (10**-2*y)-(10**-8*y**2)
#initialization
xi=1980;xf=2020;h=5
n=int((xf-xi)/h)
x=1980;y=10**5
print('x   \t\t y \t\t f(x) \t\t |f(x)-y| ')
print('%f \t %f \t %f \t  %f'%(x,y,f(x),abs(f(x)-y)))
x_plot=[];y_euler=[];y_exact=[]
#Eulers Method
for i in range(1,n+1):
     x_plot.append(x);y_euler.append(y);y_exact.append(f(x))
     y= y+h*dy(x,y)
     x=x+h
     print('%f \t %f \t %f \t %f'%(x,y,f(x),abs(f(x)-y)))
x_plot.append(x);y_euler.append(y);y_exact.append(f(x))
plt.plot(x_plot,y_euler,'bo-',x_plot,y_exact,'ro-')
plt.legend(['Numerical Solution(Euler Method)','Exact Solution'])
plt.xlabel('Time (Year)');plt.ylabel('Population')
plt.grid()
plt.savefig('Euler_logistic.png')

safa_euler=y_euler
```
