```python
%matplotlib notebook
import matplotlib.pyplot as plt
import math
#function
dy=lambda x,y: -0.06*y
f=lambda x:10**7*math.exp(-3*x/50)
#initialization
xi=0;xf=7;h=1
n =int((xf-xi)/h)
x=0;y=10**7
#Visualization of code initialization
print('x   \t\t\t y \t\t\t f(x) \t\t |f(x)-y|')
print('%f \t %f \t %f \t %f'%(x,y,f(x),abs(f(x)-y)))
x_plot=[];y_rk4=[];y_exact=[]
#RK4 Method
for i in range(1,n+1):
    x_plot.append(x);y_rk4.append(y);y_exact.append(f(x))
    #rk4 method
    k1=h*dy(x,y)
    k2=h*dy(x+h/2,y+k1*0.5)
    k3=h*dy(x+h/2,y+k2*0.5)
    k4=h*dy(x+h,y+k3)
    #Calculate new y estimation
    y=y + 1 / 6 * (k1+2*k2+2*k3+k4)
    #Increase x by step size to calculate next y extimate
    x=x+h
    print('%f \t %f \t %f \t %f'%(x,y,f(x),abs(f(x)-y)))
    
#visulation code
x_plot.append(x);y_rk4.append(y);y_exact.append(f(x))
plt.plot(x_plot,y_rk4,'go-',x_plot,y_exact,'r--')
plt.xlabel('Time (Week)');plt.ylabel('Concentration of Bacteria (parts/m^3)')
plt.legend(["Numerical Solution(RK4 Method)","Exact solution"])
safa_rk=y_rk4
safa_exact=y_exact
plt.grid()
plt.show()
plt.savefig('Rk4_Bact.png')
```
