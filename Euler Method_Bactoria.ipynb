```python
#function
%matplotlib notebook
import math
import matplotlib.pyplot as plt
f=lambda x:10**7*math.exp(-3*x/50)
dy=lambda x,y: -0.06*y
#initialization
xi=0;xf=7;h=1
n=int((xf-xi)/h)
x=0.0;y=10**7
print('x   \t\t\t y \t\t\t f(x) \t\t |f(x)-y| ')
print('%f \t %f \t %f \t %f'%(x,y,f(x),abs(f(x)-y)))
x_plot=[];y_euler=[];y_exact=[]
#Eulers Method
for i in range(1,n+1):
     x_plot.append(x);y_euler.append(y);y_exact.append(f(x))
     y= y+h*dy(x,y)
     x=x+h
     print('%f \t %f \t %f \t %f'%(x,y,f(x),abs(f(x)-y)))
x_plot.append(x);y_euler.append(y);y_exact.append(f(x))
plt.plot(x_plot,y_euler,'bo-',x_plot,y_exact,'ro-')
plt.legend(['Numerical Solution(Euler Method)','Exact solution'])
plt.xlabel('Time (Week)');plt.ylabel('Concentration of Bacteria (parts/m^3)')
plt.grid()
#plt.title('Euler Method vs Exact Solution')
safa_euler=y_euler
plt.savefig('Euler_Bact.png')
```
