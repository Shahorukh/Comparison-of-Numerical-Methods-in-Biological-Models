```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# define the sytem of differential equation in python

```python
def cde(var,t):
    x=var[0]
    y=var[1]
    dxdt=x*(0.1-0.0014*y)
    dydt=y*(0.08*x-0.0012)
    return [dxdt,dydt]
```

```python
var0=[4,9]
```

```python
cde(var0,0)
```

```python
t=np.arange(0,16,1)
```

```python
t
```

```python
solution=odeint(cde,var0,t)
```

```python
solution
```

```python
plt.plot(t,solution[:,0],'ro-')
plt.plot(t,solution[:,1],'go-')
plt.xlabel("time (Day)")
plt.ylabel("Prey and predator populaton")
plt.legend(['Prey population','predator population'])
plt.grid()
plt.show()
plt.savefig('Prey_predator_exact.png')
```
