```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.arange(1980,2025,5)
```

```python
t
```

```python
y=(10**6)/(1+9*np.exp(19.8-t/100))
```

```python
plt.plot(t,y,'ro-')
plt.legend(['Exact Solution'])
plt.xlabel('Time (Week)');plt.ylabel('Population')
plt.grid()
plt.show()
plt.savefig('Exact_logis.png')
```
